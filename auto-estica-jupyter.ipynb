```python
os.chdir('/Users/marceloalvesferreira/Desktop/recipes2/biphenylene/auto-strain/1%/DOS')

OldInput = []

with open('dftb_in.hsd' , 'r') as f:
    for line in f.readlines():
        NewLine = line.replace("\n" , "")
        OldInput.append(NewLine)

print(OldInput)

NewInput = []

for i in range(52):
    if i != 10 and i != 11:
        NewInput.append(OldInput[i])
    elif i == 10:
        NewInput.append(OldInput[10].replace("#" , ""))
    else:
        NewInput.append(OldInput[11].replace("#" , ""))

print(NewInput)

with open('dftb_in.hsd' , 'w') as g:
    for line in NewInput:
        g.write(line)
        g.write("\n")
```
